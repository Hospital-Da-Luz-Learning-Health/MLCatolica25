```python
import pandas as pd
```

```python
df = pd.read_csv('data/heart_disease.csv')
```

```python
# preprocessing
df['sex'] = df['sex'].replace({1:"male",0:'female'})
df['cp'] = df['cp'].replace({1: 'typical angina', 2:"atypical angina", 3: 'non-anginal pain', 4:'asymptomatic'})
df['restecg'] = df['restecg'].replace({0:'normal', 1:'ST-T wave abnormality', 2: 'left ventricular hypertrophy'})
df['slope'] = df['slope'].replace({1:'upsloping', 2:'fat',3:'downsloping'})
df['thal'] = df['thal'].replace({3:'normal', 6:'fixed defect', 7:'reversible defect'})
df['Diagnosed'] = df['Diagnosed'].replace({0:'<50%', 1:'>50%'})
```

```python
df
```

```python
df = df.reset_index()
df = df.rename(columns={'index':'id'})
df = df.sample(frac=1)
df
```

```python
df.to_csv('data/heart_disease_cleaned.csv',index=False)
```
